# Credit Card Approval Prediction Exploratory Data Analysis (EDA)

## Overview
### Exploratory data analysis (EDA) is a crucial step in understanding the structure and characteristics of a dataset. In this project, we conduct EDA on a dataset containing information about credit card applicants. The goal is to gain insights into various factors that may influence credit card approval decisions.

```python
# Merge Both dataset to new dataset which is "record"
record = pd.merge(App_rec,Credit_rec,on='ID',how= 'left')
record
```

```python
# find null values
record.isna().sum()
```

```python
# drop null values
record.dropna(inplace = True)
```

```python
record.isna().sum()
```

```python
# check duplicate entry
record['ID'].duplicated().sum()
```

```python
# drop duplicate entry
record.drop_duplicates(inplace = True)
```

```python
# finding count,mean,standard deviation,minimum,25% percentile,50% percentile,75% percentile, Maximun
record.describe()
```

```python
# Dataset Info
record.info()
```

```python
# finding mean value of all the income type on the bases of total income 
type = []
income = []
for category in record["NAME_INCOME_TYPE"].unique():
    x = record[record["NAME_INCOME_TYPE"] == category]['AMT_INCOME_TOTAL'].mean()
    print(f"Mean income for {category}: {round(x, 2)}")
    type.append(category)
    income.append(round(x, 2))
```

```python
# creating graphical representation
colors = ['blue', 'green', 'red', 'purple','yellow']
plt.bar(type,income,color= colors)
plt.xticks(rotation=90)
plt.title('Mean of income Base on type')
plt.xlabel('Type')
plt.ylabel('Income')
```

```python
# creating graphical representation of all the columns which have obect datatype using seaborn countplot
for i in record.columns:
    if record[i].dtype == 'object':
        sns.countplot(data=record, x=i)
        plt.xticks(rotation=90)
        plt.title(f'Count Plot for {i}')
        plt.show()
```

```python
# creating graphical representation of Months balance of both genders on the base of income type 
sns.scatterplot(x="NAME_INCOME_TYPE",y="MONTHS_BALANCE",hue="CODE_GENDER",data=record)
plt.xlabel('INCOME_TYPE')
plt.ylabel('MONTHS_BALANCE')
plt.title('Months balance of both genders on the base of income type')
plt.xticks(rotation=90)
```

```python
# finding mean of all the flag on the bases of different type of Occupation
Occupation = []
All_flag = []

for occupation_type in record["OCCUPATION_TYPE"].unique():
    subset = record[record["OCCUPATION_TYPE"] == occupation_type]
    mean_flags = subset[["FLAG_MOBIL", "FLAG_WORK_PHONE", "FLAG_EMAIL"]].mean()
    Occupation.append(occupation_type)
    All_flag.append(round(mean_flags, 2))
    print(f"All_over_Progress--{occupation_type}: {round(mean_flags, 2)}")
```

```python
# creating graphical representation
plt.figure(figsize=(10, 6))
plt.bar(Occupation, [flag["FLAG_MOBIL"] for flag in All_flag], label="FLAG_MOBIL",color = "#DE3163")
plt.bar(Occupation, [flag["FLAG_WORK_PHONE"] for flag in All_flag], bottom=[flag["FLAG_MOBIL"] for flag in All_flag], label="FLAG_WORK_PHONE",color = "#6495ED")
plt.bar(Occupation, [flag["FLAG_EMAIL"] for flag in All_flag], bottom=[flag["FLAG_MOBIL"] + flag["FLAG_WORK_PHONE"] for flag in All_flag], label="FLAG_EMAIL",color="#DFFF00" )
plt.xlabel('Occupation Type')
plt.ylabel('Mean Value')
plt.title('Mean Values of Flags by Occupation Type')
plt.xticks(rotation=90)
plt.legend()
plt.tight_layout()
```

```python
Housing_type =[]
Family_member = []
for i in record["NAME_HOUSING_TYPE"].unique():
    v = record[record["NAME_HOUSING_TYPE"] == i]["CNT_FAM_MEMBERS"].mean()
    print(f"Total_number_of_type--{i}:{round(v,2)}")
    Housing_type.append(i)
    Family_member.append(round(v,2))
```

```python
colors=["#C39BD3","#CB4335","#3498DB","#27AE60","#E67E22","#7F8C8D"]
plt.bar(Housing_type,Family_member,color=colors)
plt.xticks(rotation=90)
plt.xlabel("Housing_type")
plt.ylabel("Family_member")
plt.title("Total family member")
```

```python
Occupation=[]
total_income = []
for i in record["OCCUPATION_TYPE"].unique():
    z = record[record["OCCUPATION_TYPE"] == i]["AMT_INCOME_TOTAL"].mean()
    print(f"Total_income--{i}:{round(z,2)}")
    Occupation.append(i)
    total_income.append(round(z,2))
```

```python
colors=["#CCFF00","#CC6600","#CC0000","#666600","#66CC00","#003300","#0288D1","#009999","#FF00FF","#FFCCCC","#BDBDBD","#3E2723",
        "#43A047","#43A047","#D32F2F","#37474F","#F57C00","#311B92"]
plt.bar(Occupation,total_income,color=colors)
plt.xticks(rotation=90)
plt.xlabel("Occupation")
plt.ylabel("total_income")
plt.title("Overall_income")
```
